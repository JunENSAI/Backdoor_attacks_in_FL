This notebook was made to understand whare are we dealing with the **GTSRB** dataset. So we need to make some Explanatory Data Analysis.

```python
from torchvision import datasets
from torchvision import transforms
import torch
import matplotlib.pyplot as plt
import numpy as np

transform = transforms.ToTensor()

trainset = datasets.GTSRB(
    root="data",
    split="train",
    download=True,
    transform=transform
)

testset = datasets.GTSRB(
    root="data",
    split="test",
    download=True,
    transform=transform
)

print(len(testset), " images in the test set")
print(len(trainset), " images in the test set")
```

```python
img, label = trainset[0]

print("Images shape:", img.shape)
print("Images dtypes :", img.dtype)
print("Labels types :", type(label))
```

```python
class_counts = torch.zeros(43, dtype=torch.int32)

for _, label in trainset:
    class_counts[label] += 1

plt.figure(figsize=(12,4))
plt.bar(range(43), class_counts)
plt.xlabel("Class index (0â€“42)")
plt.ylabel("Number of images")
plt.title("GTSRB Training Set Class Distribution")
plt.show()
```

```python
import random
plt.figure(figsize=(10,3))

for i in range(5):

    j = random.randint(0, len(trainset) - 1)
    img, label = trainset[j]
    img = img.permute(1, 2, 0).numpy()
    
    plt.subplot(1, 5, i+1)
    plt.imshow(img)
    plt.title(f"Class: {label}")
    plt.axis("off")

plt.show()
```
